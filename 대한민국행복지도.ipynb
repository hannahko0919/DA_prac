```python
import pandas as pd

happy_life=pd.read_excel('data/대한민국 삶의 만족도.xlsx',engine='openpyxl')
life_merge=happy_life.groupby('시도')[['삶의 만족도']].mean()
# happy_health.head()
```

```python
happy_health=pd.read_excel('data/대한민국 건강.xlsx',engine='openpyxl')
health=happy_health.groupby('시도')[['평균']].mean()
life_merge=pd.merge(life_merge,health.rename(columns={'평균':'건강'}),on='시도')
life_merge
```

```python
import os

path='data/'
file_list=os.listdir(path)

file_list
```

```python
keywords={'건강':'happy_health','안전':'happy_safe','환경':'happy_environ','경제':'happy_econo',
          '교육':'happy_edu','관계및사회참여':'happy_relation','여가':'happy_leisure'}


try:
    for i in file_list:
        if i.endswith('.xlsx'):
            for x,y in keywords.item():
                if x in i:
                    keywords[x]=pd.read_excel(path+i,engine='openpyxl')
                    health=keywords[x].groupby('시도')[['평균']].mean()
                    happy_merge=pd.merge(happy_merge,health.rename(columns={'평균':x}),on='시도')
except:
    print(i)
```
